```python
from pyspark import SparkConf, SparkContext
conf = (SparkConf()
         .setMaster("local")
         .setAppName("Lab3")
         .set("spark.executor.memory", "1g"))
sc = SparkContext(conf = conf)
```

```python
import random
def inside(p):
    x, y = random.random(), random.random()
    return x*x + y*y < 1

count = sc.parallelize(xrange(0, 10000000)).filter(inside).count()
print ("Pi is roughly %f" % (4.0 * count / 10000000))
```

```python
import random

random.seed(1)

def sample(p):
    x, y = random.random(), random.random()
    return 1 if x*x + y*y < 1 else 0

count = sc.parallelize(range(0, 10000000)).map(sample).reduce(lambda a, b: a + b)

print("Pi is (very) roughly {}".format(4.0 * count / 10000000))

sc.stop()
```

```python
from __future__ import print_function

import sys
from operator import add

from pyspark.sql import SparkSession


if __name__ == "__main__":
    spark = SparkSession.builder.appName("PythonWordCount").getOrCreate()

    lines = spark.read.text("input/input.txt").rdd.map(lambda r: r[0])
    counts = lines.flatMap(lambda x: x.split(' ')).map(lambda x: (x, 1)).reduceByKey(add)
    output = counts.collect()
    for (word, count) in output:
        print("%s: %i" % (word, count))

    spark.stop()
```

```python
from pyspark import SparkConf, SparkContext
import os
import shutil

conf = (SparkConf()
         .setMaster("local")
         .setAppName("Lab3")
         .set("spark.executor.memory", "1g"))
sc = SparkContext(conf = conf)

if(os.path.exists("mydata")):
    shutil.rmtree("mydata")

data = [1,3,4,5,6]
distdata = sc.parallelize(data)
rdd = sc.parallelize(range(1,4)).map(lambda x: (x, "a" * x))
rdd.saveAsSequenceFile("mydata")
print(sorted(sc.sequenceFile("mydata").collect()))
sc.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
import os

spark = SparkSession.builder.appName("Sample Examples").getOrCreate()
spark_home = os.environ['SPARK_HOME']
textFile = spark.read.text(spark_home + "/README.md")
print(textFile.count())
print(textFile.first())
linesWithSpark = textFile.filter(textFile.value.contains("Spark"))
textFile.filter(textFile.value.contains("Spark")).count()
textFile.select(size(split(textFile.value, "\s+")).name("numWords")).agg(max(col("numWords"))).collect()
wordCounts = textFile.select(explode(split(textFile.value, "\s+")).alias("word")).groupBy("word").count()
print(wordCounts.collect())
linesWithSpark.cache()
print(linesWithSpark.count())
spark.stop()
```
